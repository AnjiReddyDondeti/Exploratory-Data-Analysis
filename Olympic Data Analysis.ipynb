```python
all_medals = summer_copy.groupby(['Country', 'Medal'])['Sport'].count().reset_index().sort_values(by = 'Sport', ascending = False)
all_medals.set_index('Country', inplace = True)
```

```python
all_country_gold = all_medals[all_medals['Medal'] == 'Gold']
all_country_silver = all_medals[all_medals['Medal'] == 'Silver']
all_country_bronze = all_medals[all_medals['Medal'] == 'Bronze']
```

```python
all_country_gold
```

```python
fig, ax = plt.subplots(1, 3, figsize=(35, 10))

ax[0].bar(all_country_gold.index[:20],all_country_gold['Sport'].values[:20])
ax[0].set_xlabel('Country')
ax[0].set_ylabel('Gold medals count')

ax[1].bar(all_country_silver.index[:20],all_country_silver['Sport'].values[:20])
ax[1].set_xlabel('Country')
ax[1].set_ylabel('silver medals count')

ax[2].bar(all_country_bronze.index[:20],all_country_bronze['Sport'].values[:20])
ax[2].set_xlabel('Country')
ax[2].set_ylabel('bronze medals count')
plt.show()
```

```python
best_male_athletes = summer_copy.groupby(['Gender', 'Athlete', 'Medal'])['Sport'].count().loc['Men'].reset_index().sort_values(by = 'Sport', ascending = False)
best_male_athletes
```

```python
best_male_athletes[best_male_athletes['Athlete'].duplicated()]
```

```python
top_10 = best_male_athletes['Athlete'].value_counts()[:10].index
top_10
```

```python
best_male_athletes.groupby('Athlete').get_group(' Michael Phelps')
```

```python
top_10_athletes = best_male_athletes[best_male_athletes['Athlete'].isin(top_10)]
top_10_athletes
```

```python
plt.bar(top_10_athletes['Athlete'], top_10_athletes['Sport'])
```

```python
sns.barplot(data = top_10_athletes,
             x = 'Athlete',
             y = 'Sport', hue = 'Medal')
plt.xticks(rotation = 90)
plt.ylabel()
plt.show()
```

```python
best_female_athletes = summer_copy.groupby(['Gender', 'Athlete', 'Medal'])['Sport'].count().loc['Women'].reset_index().sort_values(by = 'Sport', ascending = False)
best_female_athletes
```

```python
best_female_athletes[:15]
```

```python
best_female_athletes['Athlete'] = best_female_athletes['Athlete'].str.strip(' ')
best_female_athletes[best_female_athletes['Athlete'] == 'Larisa Latynina']
```

```python
best_female_names = best_female_athletes['Athlete'].values[:15]
best_female_names
```

```python
top_15 = best_female_athletes[best_female_athletes['Athlete'].isin(best_female_names)]
top_15
```

```python
sns.barplot(data = top_15, x = 'Athlete', y = 'Sport', hue = 'Medal')
plt.title('Best Female Athletes')
plt.xlabel('Names')
plt.ylabel('Medal Count')
plt.xticks(rotation = 90)
plt.show()
plt.style.use('fivethirtyeight')
```
