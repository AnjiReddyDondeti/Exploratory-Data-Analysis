```python
from collections import Counter
course_counter = Counter()

for course in df['LearnCodeCoursesCert'].dropna():
    course_counter.update(course.split(';'))
```

```python
course_counter.keys()
```

```python
course_counter.values()
```

```python
plt.figure(figsize = (6, 4))
plt.style.use('fivethirtyeight')
plt.bar(course_counter.keys(),
       course_counter.values())
plt.title('Most popular online learning platform')
plt.xlabel('Learning Platform')
plt.ylabel('Learners Count')
plt.xticks(rotation = 50)
plt.show()
```

```python
working_language_counter = Counter()
for language in df['LanguageHaveWorkedWith'].dropna():
    working_language_counter.update(language.split(';'))
```

```python
data_base_counter = Counter()
for data_base in df['DatabaseHaveWorkedWith'].dropna():
    data_base_counter.update(data_base.split(';'))
```

```python
want_to_work_data_base_counter = Counter()
for data_base in df['DatabaseWantToWorkWith'].dropna():
    want_to_work_data_base_counter.update(data_base.split(';'))
```

```python
fig, ax = plt.subplots(2, 2, figsize = (30, 30))

ax[0][0].barh(list(working_language_counter.keys()), list(working_language_counter.values()))
ax[0][0].set_xlabel('Language Working with')
ax[0][0].set_ylabel('Working people Count')

ax[0][1].barh(list(want_to_work_language_counter.keys()), list(want_to_work_language_counter.values()))
ax[0][1].set_xlabel('Language want to Work with')
ax[0][1].set_ylabel('people Count')

ax[1][0].barh(list(data_base_counter.keys()), list(data_base_counter.values()))
ax[1][0].set_xlabel('DataBase working with')
ax[1][0].set_ylabel('Working people Count')

ax[1][1].barh(list(want_to_work_data_base_counter.keys()), list(want_to_work_data_base_counter.values()))
ax[1][1].set_xlabel('DataBase want to work with')
ax[1][1].set_ylabel('people Count')
plt.show()
```

```python
country_group = df.groupby(['Country', 'WorkExp'])
python_df = country_group['LanguageHaveWorkedWith'].apply(lambda x : x.str.contains('Python').sum()).reset_index()
python_df.rename(columns = {'LanguageHaveWorkedWith' : 'Python_know'}, inplace = True)
python_df.sort_values(by = ['WorkExp', 'Python_know'], ascending = [False, False], inplace = True)
```

```python
java_df = country_group['LanguageHaveWorkedWith'].apply(lambda x : x.str.contains('Java').sum()).reset_index()
java_df.rename(columns = {'LanguageHaveWorkedWith' : 'Java_know'}, inplace = True)
java_df.sort_values(by = ['WorkExp', 'Java_know'], ascending = [False, False], inplace = True)
```

```python
javascript_df = country_group['LanguageHaveWorkedWith'].apply(lambda x : x.str.contains('JavaScript').sum()).reset_index()
javascript_df.rename(columns = {'LanguageHaveWorkedWith' : 'JavaScript_know'}, inplace = True)
javascript_df.sort_values(by = ['WorkExp', 'JavaScript_know'], ascending = [False, False], inplace = True)
```

```python
sql_df = country_group['LanguageHaveWorkedWith'].apply(lambda x : x.str.contains('SQL').sum()).reset_index()
sql_df.rename(columns = {'LanguageHaveWorkedWith' : 'SQL_know'}, inplace = True)
sql_df.sort_values(by = ['WorkExp', 'SQL_know'], ascending = [False, False], inplace = True)
```

```python
html_css_df = country_group['LanguageHaveWorkedWith'].apply(lambda x : x.str.contains('HTML/CSS').sum()).reset_index()
html_css_df.rename(columns = {'LanguageHaveWorkedWith' : 'HTML and CSS_know'}, inplace = True)
html_css_df.sort_values(by = ['WorkExp', 'HTML and CSS_know'], ascending = [False, False], inplace = True)
```

```python
typescript_df = country_group['LanguageHaveWorkedWith'].apply(lambda x : x.str.contains('TypeScript').sum()).reset_index()
typescript_df.rename(columns = {'LanguageHaveWorkedWith' : 'TypeScript_know'}, inplace = True)
typescript_df.sort_values(by = ['WorkExp', 'TypeScript_know'], ascending = [False, False], inplace = True)
```

```python
c_hash_df = country_group['LanguageHaveWorkedWith'].apply(lambda x : x.str.contains('C#').sum()).reset_index()
c_hash_df.rename(columns = {'LanguageHaveWorkedWith' : 'C#_know'}, inplace = True)
c_hash_df.sort_values(by = ['WorkExp', 'C#_know'], ascending = [False, False], inplace = True)
```
