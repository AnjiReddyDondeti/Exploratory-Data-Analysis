```python
country_group = df1.groupby('Country')
india = country_group.get_group('India')
india['EdLevel'].value_counts()
```

```python
india['MgrIdiot'].value_counts()
```

```python
language = india['LanguageWorkedWith'].dropna().values
language
```

```python
from collections import Counter

counter = Counter()

for response in language:
    counter.update(response.split(';'))

languages = []
people = []

for item in counter.most_common():
    languages.append(item[0])
    people.append(item[1])
print(languages)
print(people)
```

```python
plt.figure(figsize = (5, 8))
plt.style.use('fivethirtyeight')
plt.barh(languages,people)
plt.title('people and languages worked in India')
plt.xlabel('people count')
plt.ylabel('languages')
```
